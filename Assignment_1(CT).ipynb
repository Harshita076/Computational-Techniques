```python
midpoints = [20, 30, 40, 50, 60]
frequencies = [350, 375, 120, 80, 90]

# Mean
total_fx = sum([f * x for f, x in zip(frequencies, midpoints)])
total_f = sum(frequencies)
mean = total_fx / total_f

print("The mean of dataset is:", mean)
```

```python
# Median calculation
class_interval = classes = [(15, 25), (25, 35), (35, 45), (45, 55), (55, 65)]
frequencies = [350, 375, 120, 80, 90]


N = sum(frequencies)
N2 = N / 2
cumulative = []
cum = 0
for f in frequencies:
    cum += f
    cumulative.append(cum)
for i, cf in enumerate(cumulative):
    if cf >= N/2:
        median_idx = i
        break
L = class_interval[median_idx][0]   # lower boundary
h = class_interval[median_idx][1] - class_interval[median_idx][0]  # class width
f = frequencies[median_idx]          # frequency of median class
CF = cumulative [median_idx - 1] if median_idx > 0 else 0  # cumulative before median class

median = L + ((N/2 - CF) / f) * h
print("The median of dataset is:",median)
```

```python
#modal class (highest frequency)
modal_idx = frequencies.index(max(frequencies))

# Extracting the  components
L = class_interval[modal_idx][0]   # lower boundary of modal class
f1 = frequencies[modal_idx]
f0 = frequencies[modal_idx - 1] if modal_idx > 0 else 0
f2 = frequencies[modal_idx + 1] if modal_idx < len(frequencies) - 1 else 0
h = class_interval[0][1] - class_interval[0][0]

# Mode formula
mode = L + ((f1 - f0) / (2*f1 - f0 - f2)) * h

print("The mode of dataset is:", mode)
```
